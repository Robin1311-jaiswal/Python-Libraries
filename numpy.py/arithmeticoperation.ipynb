#arithmertic operation 


```python
import numpy as np
```

```python
var = np.array([1,2,3,4])
varadd= var + 3
print(varadd)
```

```python
var = np.array([1,2,3,4])
var2= np.array([5,6,7,8])
print(var+var2)
```

```python
var = np.array([4,5,6,6])
print(var-4)
```

```python
var = np.array([3,4,5,6])
varadd = var*7
print(varadd)
```

```python
var = np.array([2,4,5,6])
varadd = var/6
print(varadd)
```

```python
var = np.array([3,4,5,678])
varadd= var%6
print(varadd)
```

```python
var = np.array([1,2,3,4])
var2= np.array([5,6,7,8])
varadd = np.add(var,var2)
print(varadd)
```
