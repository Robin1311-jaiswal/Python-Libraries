```python
import numpy as np
```

```python
var = np.array([2,34,5,6])
print("min:", np.min(var), np.argmin(var))
print("max:", np.max(var), np.argmax(var))
print("sqrt:", np.sqrt(var))
```

```python
var = np.array([[2,3,4,5], [3,4,5,6]])
print(np.min(var, axis=1))
print(np.max(var, axis=0))
print(np.cos(var))
print()
print(np.sin(var))
print()
print(np.cumsum(var))
```
