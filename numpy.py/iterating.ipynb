```python
import numpy as np
```

```python
var = np.array([1,2,3,4,4,5,6,67,67])
print(var)
print()
for i in var:
    print(i)
```

```python
var = np.array([[2,3,4,5,6,7,7],[3,4,5,6,7,8,7]])
print(var)
print(var.shape)
print(var.ndim)
print()
for i in var:
    print(i)

    print()
for k in var:
    for l in k:
        print(l)
```

```python
var= np.array([[[2,3,4,5,6],[1,2,3,4,4]]])
print(var.ndim)
print()
for i in var:
    for j in i:
        for k in j:
            print(k)
```

```python
var= np.array([[[2,3,4,5,6],[1,2,3,4,4]]])
print(var.ndim)
print()
for i in np.nditer(var, flags=['buffered'], op_dtypes=["S"]):
    print(i)
```

```python
var= np.array([[[2,3,4,5,6],[1,2,3,4,4]]])
print(var.ndim)
print()
for i , d in np.ndenumerate(var):
    print(i,d)
```
