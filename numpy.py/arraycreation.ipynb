###zeros

```python
import numpy as np 

ar_zero = np.zeros(4)
ar_zero1 = np.zeros((3,4))
print(ar_zero)
print()
print(ar_zero1)
```

```python
ar_one = np.ones(4)
print(ar_one)
```

```python
ar_em = np.empty(4)
print(ar_em)
```

```python
ar_rn = np.arange(4)
print(ar_rn)
```
