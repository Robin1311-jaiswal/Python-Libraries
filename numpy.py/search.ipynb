```python
import numpy as np
```

```python
arr = np.array([1,2,3,3,4,4,5,4,3,6,7,6,5,7])
x= np.where((arr/2)==3)
print(x)
```

```python
arr = np.array([1,2,3,3,4,4,5,4,3,6,7,6,5,7])
x= np.where(arr==3)
print(x)
```
