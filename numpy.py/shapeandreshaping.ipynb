```python
import numpy as np 
```

```python
var = np.array([[1,2,3,4],[2,4,5,6]])
print(var)

print(var.ndim)
print()
print(var.dtype)
print(var.shape)
```

```python
var = np.array([1,2,3,4], ndmin=3)
print(var)

print(var.ndim)
print()
print(var.dtype)
print(var.shape)
```
