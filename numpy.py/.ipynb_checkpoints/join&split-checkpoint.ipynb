```python
import numpy as np
```

```python
var= np.array([2,3,4,5])
var1=np.array([2,5,6,7])

ar= np.concatenate((var, var1))
print(ar)
```

```python
var = np.array([[2,3,4,5], [3,4,5,6]])
var1 =np.array([[4,5,6,7],[6,7,8,9]])
ar= np.concatenate((var,var1),axis=1)
print(ar)
```

```python
var= np.array([2,3,4,5])
var1=np.array([2,5,6,7])

ar= np.stack((var,var1), axis=1)
ar2= np.hstack((var,var1)) #row
ar3 = np.vstack((var,var)) #column
ar4= np.dstack((var,var1)) #height
print()
print(ar)
print()
print(ar2)
print()
print(ar3)
print()
print(ar4)
```
