```python
import pandas as pd
```

```python
l= [1,2,3,4,5,6]
var = pd.DataFrame(l)
print(var)
print(type(var))
```

```python
d={"a": [1,2,3,4,5], "s":[1,2,3,4,5]}

var1 = pd.DataFrame(d)
print(var1)
print(type(var1))
```

```python
d ={"a": [1,2,3,4,5], "s":[1,2,3,4,5]}

var1 = pd.DataFrame(d, columns=["a"],index=["a","s","d","f","g"])
print(var1)
print(type(var1))
```
