```python
import pandas as pd
```

```python
x= [2,3,4,5]
var = pd.Series(x)
print(var)
```

```python
x=[4,5,6,7]
var1= pd.Series(x, index=['a', 's', 'd', 'f'], dtype= "float", name="python")

print(var1)
print(type(var1))
print(var1[2])
```

```python
dic = {"name" : ['python','c','c++','java'], "por": [11,1,3,12], "rank": [1,2,3,4]}
var2=pd.Series(dic)
print(var2)
```

```python
s= pd.Series(12)

print(s)
print(type(s))
```

```python
s= pd.Series(12, index=[1,2,3,4,5,5,6,7])

print(s)
print(type(s))
```

```python
s1= pd.Series(12, index=[1,2,3,4,5,6,7])
s2= pd.Series(12, index=[1,2,3,4])

print(s1+s2)
```
