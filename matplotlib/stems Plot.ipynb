```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,4,3,6]

plt.stem(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,4,3,6]

plt.stem(x,y, linefmt=":", markerfmt="r+", bottom =1) #or for markerfmt = "ro"
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,4,3,6]

plt.stem(x,y, linefmt=":", markerfmt="ro", bottom =1, basefmt="g", label="python", orientation= "horizontal")
plt.legend()
plt.show()
```
