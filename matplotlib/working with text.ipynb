```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y= [3,2,4,1,5]

plt.plot(x,y)
plt.xlabel("language",fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("progarmming", fontsize=15)

plt.text(2,5,"java", fontsize=15, style="italic", bbox={"facecolor":"red"})

plt.annotate("python",xy=(2,1),xytext=(3,3), arrowprops=dict(facecolor="black", shrink=100))

plt.legend(["up"],loc=9, facecolor="r",edgecolor="y", framealpha=0.5, shadow=True)
```
