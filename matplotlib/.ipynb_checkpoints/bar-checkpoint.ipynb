```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=["python", "C", "C++", "Java"]
y= [87,90,76,89]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Jaiswal", fontsize=15)
c=["y","b","m","g"]

plt.bar(x,y, width=0.4, color=c, edgecolor="r", label ="Jaiswal") # linewidth = 4, linestyle=":",          #        #align="center" #edge alpha value 0 to 1
plt.legend() # for using the cour the label jaiswal
plt.show()
```

```python
x=["python", "C", "C++", "Java"]
y= [87,90,76,89]
z= [20,30,40,50]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Jaiswal", fontsize=15)
c=["y","b","m","g"]

plt.bar(x,y, width=0.4, color=c,label ="Jaiswal")
plt.bar(x,z, width =0.4, color ="b", label = "jaiswal2")

plt.legend() # for using the cour the label jaiswal
plt.show()
```

```python
x=["python", "C", "C++", "Java"]
y= [87,90,76,89]
z= [20,30,40,50]

width = 0.2
p=np.arange(len(x))
p1=[j+width for j in p]


plt.xlabel("language",fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Jaiswal", fontsize=15)


plt.bar(p,y, width, color="r",label ="Jaiswal")
plt.bar(p1,z, width, color ="b", label = "jaiswal2")

plt.legend() # for using the cour the label jaiswal
plt.show()
```
