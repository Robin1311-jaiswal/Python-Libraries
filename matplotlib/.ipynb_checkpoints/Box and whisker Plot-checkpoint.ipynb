```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x, notch= True, vert= False)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x, widths=0.3, labels=["python"], patch_artist= True, showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70, 130]

plt.boxplot(x, widths=0.3, patch_artist= True, showmeans=True, sym="g", boxprops=dict(color="r"), capprops=dict(color="r"), whiskerprops=dict(color="r"))
plt.show()
```

```python
x=[10,20,30,40,50,60,70, 130]

plt.boxplot(x, widths=0.3, patch_artist= True, showmeans=True, whis=2)
plt.show()
```

```python
x=[10,20,30,40,50,60,70, 130]
x1=[10,40,30,20,50,40,50,39]
y=[x,x1]

plt.boxplot(y,labels=["python","c++"], widths=0.3,  showmeans=True, sym="g", boxprops=dict(color="r"), capprops=dict(color="r"), whiskerprops=dict(color="r"))
plt.show()
```
