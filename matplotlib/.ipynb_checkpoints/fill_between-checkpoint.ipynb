```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x=[1,2,3,4,5]
y=[3,2,4,5,4]
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,marker="o", ms=10,mfc="g", label="python", color="r")
plt.fill_between(x,y, where =(x>=2) & (x<=4), alpha=0.3)
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
```
