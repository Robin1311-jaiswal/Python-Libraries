```python
import matplotlib.pyplot as plt
```

```python
x= [10, 20, 30, 40]
y=["C", "C++", "Java", "Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","b", "y","o"]
plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%" , shadow=True, radius=1, labeldistance=1.1, startangle=120, 
       textprops={"fontsize": 15}, counterclock= True, wedgeprops={"linewidth":5}, rotatelabels= False)

plt.title("Programming Subject")
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
x= [10, 20, 30, 40]

x1=[40,50,60,44]
y=["C", "C++", "Java", "Python"]


plt.pie(x, labels=y, radius=1.5)
plt.pie(x1, radius=0.5)

plt.show()
```

```python
x= [10, 20, 30, 40]

x1=[40,50,60,44]
y=["C", "C++", "Java", "Python"]


plt.pie(x, labels=y, radius=1.5)
plt.pie([1], colors="w")

plt.show()
```

```python
x= [10, 20, 30, 40]

x1=[40,50,60,44]
y=["C", "C++", "Java", "Python"]


plt.pie(x, labels=y, radius=1.5)
cr=plt.Circle(xy=(0,0), radius=1, facecolor="w")
plt.gca().add_artist(cr)

plt.show()
```
