```python
import  matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5,6,7]
y=[2,3,1,4,5,6,4]
plt.scatter(x,y)

plt.show()
```

```python
day= [1,2,3,4,5,6,7]
No=[2,3,1,4,5,6,4]
plt.scatter(day, No, color ="r" )
plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

plt.show()
```

```python
day= [1,2,3,4,5,6,7]
No=[2,3,1,4,5,6,4]
colors=["r","y","g","b","r","g","r"]
#size=[]
plt.scatter(day, No, color = colors, marker = "*", edgecolor="g", linewidth=3 ) # alpha used for transparent graph
plt.colorbar()# line vertical colorful are colorbar
plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

plt.show()
```

```python
day= [1,2,3,4,5,6,7]
No=[2,3,1,4,5,6,4]
colors=[10,20,34,56,7,43,23]
sizes=[200,300,300,100,50,40,150]
plt.scatter(day, No, c = colors, s=sizes, cmap="viridis" ) # alpha used for transparent graph\
t=plt.colorbar()
t.set_label("colorbar")
plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

plt.show()
```
