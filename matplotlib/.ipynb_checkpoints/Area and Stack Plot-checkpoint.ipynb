```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area1= [2,4,3,5,7]
area2=[3,1,2,5,4]
area3= [5,9,5,4,3]

l=["area1","area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l)
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc=2)
plt.show()
```
