```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv("C:\\Users\\Ranjan Kumar\\Downloads\\titanic.csv")
```

```python
data_corr=dataset.select_dtypes(["float64", "int64"]).corr()
data_corr
```
