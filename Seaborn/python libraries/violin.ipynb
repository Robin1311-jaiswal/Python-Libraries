```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd 

```

```python
var= sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="time", linewidth=2, palette="Accent", saturation=2)
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=var, order=["Dinner", "Lunch"], inner="quart")#inner box point 
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="sex", split=True)
plt.show()
```
