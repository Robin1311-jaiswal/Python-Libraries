```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd 

```

```python
var= sns.load_dataset("penguins")
var
```

```python
sns.barplot(x=var.island, y=var.bill_length_mm)
plt.show()
```

```python
order_1= ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order_1,ci=90, saturation=0.5,errcolor="y", errwidth=12)  #if i dont want to use the variable name var then instead of that  i can use is as str.
plt.show()
```

```python
order_1= ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order_1,errcolor="y")  #if i dont want to use the variable name var then instead of that  i can use is as str.
plt.show()
```

```python
sns.set(style="dark")
order_1= ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order_1, dodge=False)  #if i dont want to use the variable name var then instead of that  i can use is as str.
plt.show()
```
