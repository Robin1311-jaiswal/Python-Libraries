```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
var= np.linspace(1,10,20).reshape(4,5)
var
```

```python
y={"fontsize":20,"color":"r"}
v=sns.heatmap(var, vmin=0, vmax=12, cmap="gist_heat", annot=True, annot_kws=y, linewidth=10, linecolor="y", cbar=False )
v.set(xlabel="python", ylabel="Programming")
sns.set(font_scale=3)
plt.show()
```
